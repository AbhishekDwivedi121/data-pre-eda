```python
import pandas as pd
```

```python
# Checking basic information about the dataset
# - Lists total entries, data types, and non-null values for each column
# - Helps identify columns with missing data that need cleaning
data.info()
```

```python
# Cleaning the 'price' column:
# - Removing the '₹' currency symbol and commas using str.replace()
# - Converting the cleaned strings to numeric values using pd.astype()
# - This allows us to treat 'price' as a numerical feature for analysis

df['price'] = df['price'].str.replace('₹', '')  # Remove rupee symbol
df['price'] = df['price'].str.replace(',', '')  # Remove commas
df['price'] = df['price'].astype(int)   # Convert to numeric
```

```python
# Assumption: Phones priced below ₹3000 are likely to be feature phones, not smartphones


df[df['price'] < 3000]
```

```python
# Removing feature phones by setting a price threshold
# This helps keep the dataset focused on smartphones only
df = df[df['price'] > 3000]
df.shape
```
