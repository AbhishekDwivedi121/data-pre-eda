```python

def is_memory_card_supported(card_value):
    if pd.isna(card_value):
        return np.nan
    card_value = str(card_value).lower()
    if "not supported" in card_value:
        return False
    elif "supported" in card_value or "hybrid" in card_value:
        return True
    else:
        return np.nan
```

```python
data['Memory Card Support'] = data['card'].apply(is_memory_card_supported).values
```

```python
import re
import numpy as np
import pandas as pd

def extract_card_capacity(card_value):
    if pd.isna(card_value):
        return np.nan
    
    # Look for common formats like '256GB', '1TB', '512 GB' (with or without weird spaces)
    pattern = r'(\d+)\s*[\u2009 ]?(GB|TB)'  # \u2009 is the thin space sometimes used
    
    match = re.search(pattern, card_value.upper())
    if match:
        size, unit = match.groups()
        return f"{size}{unit}"
    
    return np.nan
```

```python
data['card'].apply(extract_card_capacity)
```

```python
data.insert(24,"Extended Memory Capacity(in Tb)", data['card'].apply(extract_card_capacity))
```
