# ðŸ“Š Univariate Analysis

Univariate analysis focuses on analyzing *a single variable* at a time. It helps in understanding the *distribution, **central tendency, and **spread* of the data for one column.

---

## ðŸŽ¯ Purpose

- Understand the basic structure of the data
- Detect outliers or unusual patterns
- Identify data skewness
- Assist in selecting appropriate transformation techniques

---




```python

data['Brand'].value_counts().head(10).plot(kind='pie',autopct='%1.1f%%',colormap='plasma')
sns.set_theme(style='whitegrid')
plt.title("Market Capture By Smartphone Brands")
```

```python
sns.histplot(data=data,x=data['price'])
plt.title("Price Graph")
```

```python
data['price'].describe().plot(kind='bar')
plt.xticks(rotation = 45)
plt.title("Price Summary statistics")
```
